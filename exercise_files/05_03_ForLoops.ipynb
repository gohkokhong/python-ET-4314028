### Continue

```python
for letter, animals in animalLookup.items():
    if len(animals) > 1:
        continue
    print(f'Only one animal! {animals}')
```

### Break

```python
for letter, animals in animalLookup.items():
    if len(animals) > 1:
        print(f'Found {len(animals)} animals: {animals}')
        break
        
    print(f'Only one animal! {animals}')
```

### For / Else

```python
for number in range(2, 100):
    for factor in range(2, int(number**0.5) + 1):
        if number % factor == 0:
            break
    else:
        print(f'{number} is prime!')
```

```python
for number in range(2,100):
    foundFactors = False

    for factor in range(2, int(number**0.5) + 1):
        if number % factor == 0:
            foundFactors = True
            break
            
    if not foundFactors:
        print(f'{number} is prime!')
```
