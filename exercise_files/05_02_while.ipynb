### Pass

```python
waitUntil = (datetime.now().second + 2) % 60

while datetime.now().second != waitUntil:
    pass

print(f'We are at {waitUntil} seconds!')
```

### Break

```python
waitUntil = (datetime.now().second + 2) % 60

while True:
    if datetime.now().second == waitUntil:
        print(f'We are at {waitUntil} seconds!')
        break
```

### Continue

Skips any lines that follow it, within the same loop

```python
waitUntil = (datetime.now().second + 2) % 60

while datetime.now().second != waitUntil:
    continue
    print('Still waiting')

print(f'We are at {waitUntil} seconds!')
```

```python
waitUntil = (datetime.now().second + 2) % 60

while True:
    if datetime.now().second < waitUntil:
        continue
    break

print(f'We are at {waitUntil} seconds!')
```
