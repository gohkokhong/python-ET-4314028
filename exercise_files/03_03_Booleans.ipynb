# Boolean logic

```python
weatherIsNice = False
haveUmbrella = True
```

```python
if not (haveUmbrella or weatherIsNice):
    print('Stay inside')
else:
    print('Go for a walk!')
```

```python
if not haveUmbrella and not weatherIsNice:
    print('Stay inside')
else:
    print('Go for a walk!')
```

```python
if haveUmbrella or weatherIsNice:
    print('Go for a walk!')
else:
    print('Stay inside!')
```
