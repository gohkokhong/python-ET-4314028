## If statements with "FizzBuzz"

```python
# 1, 2, Fizz, Buzz, Fizz, 7, 8, Fizz, Buzz, 11, Fizz, 13, FizzBuzz, 16

# Divisible by 3 = Fizz
# Divisible by 5 = Buzz
# Divisible by 15 = FizzBuzz
# Otherwise, print the number

for n in range(0, 101):
    if n % 15 == 0:
        print('FizzBuzz')
    elif n % 3 == 0:
        print('Fizz')
    elif n % 5 == 0:
        print('Buzz')
    else:
        print(n)
```
