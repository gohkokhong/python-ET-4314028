## Dictionary Comprehensions

```python
animalList = [
    ('a', 'aardvark'),
    ('b', 'bear'),
    ('c', 'cat'),
    ('d', 'dog'),
]

animals = {item[0]: item[1] for item in animalList}
animals
```

```python
animals = {key: value for key, value in animalList}
animals
```

```python
animals.items()
```

```python
list(animals.items())
```

```python
[{'letter': key, 'name': value} for key, value in animals.items()]
```
