### If / Else statements

```python
a = True

if a:
    print("It is true!")
    print("also print this")
else:
    print("It is false!")

print("Always print this")
```

## Loops

```python
a = [1,2,3,4,5]

for item in a:
    print(item)
```

```python
a = 0

while a < 5:
    print(a)
    a += 1
```
