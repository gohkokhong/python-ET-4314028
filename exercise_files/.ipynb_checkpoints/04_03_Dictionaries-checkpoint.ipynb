## Dictionaries

```python
animals = {
    'a': 'aardvark',
    'b': 'bear',
    'c': 'cat',
}

animals
```

```python
animals['a']
```

```python
animals['d'] = 'dog'
```

```python
animals
```

```python
animals['a'] = 'antelope'
animals
```

```python
animals.keys()
```

```python
animals.values()
```

```python
list(animals.keys())
```

```python
animals['e']
```

```python
animals.get('e', '')
```

```python
print(animals.get('e'))
```

```python
animals.get('a')
```

```python
len(animals)
```

```python
animals = {
    'a': ['aardvark', 'antelope'],
    'b': ['bear'],
}

animals['b'].append('bison')
```

```python
if 'c' not in animals:
    animals['c'] = []

animals['c'].append('cat')
```
