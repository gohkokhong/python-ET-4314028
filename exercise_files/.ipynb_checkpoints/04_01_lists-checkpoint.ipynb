## Lists

### List Slicing

```python
myList = [1,2,3,4,5]
myList[3:]
```

```python
myList[0:6:2]
```

```python
myList[0:6:3]
```

```python
myList[::2]
```

```python
for i in range(100):
    print(i, end=' ')
```

```python
myList = list(range(100))
myList[::10]
```

```python
myList[::-10]
```

### Modifying Lists

```python
myList = [1,2,3,4]
myList.append(5)
print(myList)
```

```python
myList.insert(3, "A new value")
print(myList)
```

```python
myList.remove('A new value')
print(myList)
```

```python
myList.pop()
print(myList)
```

```python
while len(myList):
    print(myList.pop())
```

```python
myList
```

```python
a = [1,2,3,4,5]
b = a
a.append(6)
print(b)
```

```python
a = [1,2,3,4,5]
b = a.copy()
a.append(6)
print(a)
print(b)
```
